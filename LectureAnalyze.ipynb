```python
import warnings
warnings.filterwarnings("ignore")
from pandas import *

data1 = read_html("http://abur29.github.io/vlad.html", header = 0)
vlad = data1[0]
data1 = read_html("http://abur29.github.io/msk.html", header = 0)
msk = data1[0]
data1 = read_html("http://abur29.github.io/maha.html", header = 0)
maha = data1[0]
data1 = read_html("http://abur29.github.io/oren.html", header = 0)
oren = data1[0]
data1 = read_html("http://abur29.github.io/spb.html", header = 0)
spb = data1[0]

table = concat([vlad,msk,maha,oren,spb])
table
```

```python
table.drop(["STATION_ID", "Q", "Q.1", "Q.2", "Q.3","D"], inplace=True, axis=1)
table.head()
```

```python
table["STATION_NM"][table["STATION_NM"]=="ВЛАДИВОСТОК"] = 1 
table["STATION_NM"][table["STATION_NM"]=="МОСКВА ВДНХ"] = 2
table["STATION_NM"][table["STATION_NM"]=="МАХАЧКАЛА"] = 3
table["STATION_NM"][table["STATION_NM"]=="ОРЕНБУРГ"] = 4
table["STATION_NM"][table["STATION_NM"]=="САНКТ-ПЕТЕРБУРГ"] = 5
table.tail()
```
